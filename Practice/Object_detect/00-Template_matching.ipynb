```python
full = cv2.cvtColor(cv2.imread('D:\ANSON\OpenCV\Computer-Vision-with-Python\DATA\\sammy.jpg'),cv2.COLOR_BGR2RGB)

face = cv2.cvtColor(cv2.imread('D:\ANSON\OpenCV\Computer-Vision-with-Python\DATA\\sammy_face.jpg'),cv2.COLOR_BGR2RGB)
plt.imshow(full)
```

```python
# All the 6 methods for comparison in a list
# Note how we are using strings, later on we'll use the eval() function to convert to function
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
height, width, channels = face.shape
```

```python
for m in methods:
    method = eval(m)

    # Create a copy from full image
    full_copy = full.copy()

    # Template matching
    res = cv2.matchTemplate(full_copy, face, method)

    # Grab the Max and Min values, plus their locations
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

    # Top left point
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc    
    else:
        top_left = max_loc

    # Bottom right point
    bottom_right = (top_left[0]+width, top_left[1]+height)

    # Draw a rectangle
    cv2.rectangle(full_copy, top_left, bottom_right, (124,252,0), 10)

    # Plot the Images
    plt.subplot(121)
    plt.imshow(res)
    plt.title('Result of Template Matching')
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('Detected Point')
    plt.suptitle(m)
    
    
    plt.show()
    print('\n')
    print('\n')
```
