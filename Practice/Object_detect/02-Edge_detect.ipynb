```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('D:\ANSON\OpenCV\Computer-Vision-with-Python\DATA\\sammy_face.jpg')
plt.imshow(img)
```

```python
blurred = cv2.GaussianBlur(img, ksize=(5,5), sigmaX=10)
plt.imshow(blurred)
```

```python
mean_val = np.median(img)
mean_val
```

```python
lower = int(max(0, 0.7*mean_val))
upper = int(min(255, 1.3*mean_val))
```

```python
edges = cv2.Canny(blurred, threshold1=lower, threshold2=upper+50)
plt.imshow(edges)
```
