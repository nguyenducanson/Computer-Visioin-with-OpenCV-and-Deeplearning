```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('D:\ANSON\OpenCV\Computer-Vision-with-Python\DATA\sudoku.jpg',0)
display_img(img)
```

```python
thres = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,7,8)
display_img(thres)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
openning = cv2.morphologyEx(thres, op=cv2.MORPH_OPEN, kernel=kernel)
display_img(openning)
```

## Sobel

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
display_img(sobely)
```

```python
blended = cv2.addWeighted(src1=sobelx, alpha=0.5, src2=sobely, beta=0.5, gamma=0)
display_img(blended)
```

```python
gradient = cv2.morphologyEx(blended, cv2.MORPH_GRADIENT, kernel)
display_img(gradient)
```

## Laplacian

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F, ksize=5)
display_img(laplacian)
```
