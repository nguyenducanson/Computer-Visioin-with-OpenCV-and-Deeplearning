```python
img2 = cv2.imread('D:\ANSON\OpenCV\Computer-Vision-with-Python\DATA\crossword.jpg',0)
plt.imshow(img2, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img2)
```

```python
ret, th1 = cv2.threshold(img2, 180, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

## Adaptive Thresholding (phn ngng thch nghi)
#### Phng thc ny tnh gi tr trung bnh ca cc n im xung quanh pixel  ri tr cho C ch khng ly ngng c nh (n thng l s l, cn C l s nguyn bt k).
### Ngoi cc tham s ging nh phn ngng thng, adaptiveThreshold c thm cc tham s:
#### Phng thc:
* ADAPTIVE_THRESH_MEAN_C: gi tr ca pixel ph thuc vo cc pixel ln cn
* ADAPTIVE_THRESH_GAUSSIAN_C: gi tr ca pixel cng ph thuc vo cc pixel ln cn, tuy nhin c kh nhiu
* Block Size: s pixel ln cn dng  tnh ton
* C: hng s tr i gi tr trung bnh

```python
th2 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 13, 8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1=th1, alpha=0.4, src2=th2, beta=0.6, gamma=0)
show_pic(blended)
```
